```python
import nltk
from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy as nltk_acc
from nltk.corpus import stopwords
import random
```

```python
nltk.download('movie_reviews')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def extract_features(word_list):
    return dict([(word, True) for word in word_list])
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
```

```python
random.shuffle(documents)
```

```python
features = [(extract_features(doc), pos) for (doc, pos) in documents]
```

```python
train_set = documents[:1600]
test_set = documents[1600:]
```

```python
# Extract features from the training set
train_features = [(extract_features(doc), pos) for doc, pos in train_set]
# Train the classifier using the extracted features
Classifier = NaiveBayesClassifier.train(train_features)
```

```python
train_features[0]
```

```python
test_features = [(extract_features(doc), pos) for doc, pos in test_set] #This line was added to fix the error
```

```python
acc = nltk_acc(Classifier, test_features)
print("Accuracy:", acc)
```
